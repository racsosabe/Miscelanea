## Fundamentos: Demostraciones

### Demostración y Axioma

Una demostración es una manera de asignar veracidad a una proposición basándose en **axiomas** predefinidos en el campo que se analice.

Un axioma es una proposición que asume veracidad sin necesidad de prueba alguna.

Por ejemplo, para la geometría Euclidiana, la proposición "_Dados dos puntos, se puede trazar una única recta que los una_" es un axioma verdadero; sin embargo, para otras geometrías, no necesariamente se cumple lo mismo. 

### Demostración por casos

Cuando se puede separar por casos una proposición y analizarlos todos, se puede llegar a una conclusión de veracidad. Es una técnica muy útil, sobretodo cuando la cantidad de casos es pequeña.

__Ejemplo:__

Dado $a, b$, con $a, b > 0$, reduce $2$ veces en $1$ las variables, pruebe que el valor mínimo del producto se obtiene solo reduciendo al menor de ellos.

Supongamos, sin perdida de generalidad que $a \le b$, tenemos los siguientes casos:

|_a_|_b_|producto|
----|----|----|
|$$a-1$$|$$b-1$$|$$ab - a - b + 1$$| 
|$$a-2$$|$$b$$  |$$ab - 2b$$|
|$$a$$| $$b-2$$ |$$ab - 2a$$|

Por lo que vemos

$$
\begin{align*}
ab - a - b + 1 &\ge ab - 2b\\
ab - 2a &\ge ab - 2b
\end{align*}$$

Concluyendo que debemos reducir $a = \min(a, b)$, dos veces.



### Demostración de implicancias

Cuando la proposición es de la forma $P \longrightarrow Q$ solo tenemos que analizar 2 casos: 

1) $P$ es falsa, en cuyo caso la proposición es verdadera, dado que $F \longrightarrow Q$ es verdadera sin importar qué valor tome $Q$. 

2) $P$ es verdadera, en cuyo caso tenemos algunas opciones para la demostración:

#### Prueba directa

Podemos asumir que $P$ es verdadero y realizar una serie de pasos lógicos para llegar a la conclusión de que $Q$ también es verdadero.

Ejemplo:

$p \vert n \text{, p es primo} \longrightarrow p \vert n^{k}, \forall k \in \mathbb{N}$

Si asumimos que $p | n$ y $p$ es primo, entonces por el teorema fundamental de la aritmética, podemos expresar a $n$:

$$ n = p^{a} \cdot \prod_{i=1}^{k}p_{i}^{a_{i}}\text{, } p_{i} \neq p \text{, } \forall i = 1,2,\ldots,k \text{ y } a > 0 $$

Además de ello, dado que $k$ es un número natural, entonces:

$$ k \geq 1 \rightarrow k-1 \geq 0 $$

Lo que significa que:

$$ n^{k} = n \cdot n^{k-1} = p^{a} \cdot \prod_{i=1}^{k}p_{i}^{a_{i}} \cdot n^{k-1} \rightarrow p | n^{k} $$

Donde $n^{k-1}$ también es un número natural debido a que $k-1$ es no negativo.

Con esto concluimos que $p$ es un factor de $n^{k}$, así la proposición es verdadera.

#### Prueba por contraposición

Podemos usar la equivalencia lógica de $$P \longrightarrow Q \equiv \sim Q \longrightarrow \sim P$$ 
Asumir que $\sim Q$ es verdadero y aplicar el paso anterior.

Ejemplo:

$$ \text{Si }r\text{ es irracional} \longrightarrow \sqrt{r}\text{ es irracional} $$

Su contraposición será:

$$ \text{Si }\sqrt{r}\text{ es racional} \longrightarrow r\text{ es racional} $$

Si $\sqrt{r}$ es racional, quiere decir que existen dos enteros $a$ y $b \neq 0$ tales que:

$$ \sqrt{r} = \frac{a}{b} $$

Si elevamos al cuadrado ambos lados de la ecuación, obtenemos:

$$ r = \frac{a^{2}}{b^{2}} $$

Y dado que $a^{2}$ y $b^{2}$ también son enteros, $r$ es racional.


### Demostración de implicancia mutua

Cuando la proposición es de la forma $P \longleftrightarrow Q$, se puede probar la equivalente lógica $P \longrightarrow Q \wedge Q \longrightarrow P$ o realizar una cadena de implicancias mutuas que conlleven a $Q$ desde $P$ o viceversa.

Ejemplo de cadena:

$$ \text{La desviación estandar de }n\text{ valores es 0} \longleftrightarrow \text{Todos los valores son iguales a la media} $$

Sean los valores $x_{i}$, para $i = 1,2,\ldots,n$, entonces la desviación estándar es:

$$ \sqrt{\frac{\sum\limits_{i=1}^{n}(x_{i}-\mu)^{2}}{n}} = 0 $$

Donde $\mu$ es la media de los $n$ valores.

Dado que 0 es el único número cuya raíz cuadrada es 0, la igualdad se cumple si y solo si:

$$ \sum\limits_{i=1}^{n}(x_{i}-\mu)^{2} = 0 $$

Dado que los cuadrados perfectos son no negativos y la suma del lado izquierdo de la ecuación es no negativa, la ecuación se cumple si y solo si:

$$ \text{Cada término de la suma del lado izquierdo es 0} $$

Tenemos ahora que:

$$ (x_{i}-\mu)^{2} = 0\text{, }\forall i=1,2,\ldots,n $$

Pero el término $(x_{i}-\mu)^{2}$ es 0 si y solo si $x_{i}-\mu$ es 0, por lo que:

$$ x_{i} -\mu = 0\text{, }\forall i=1,2,\ldots,n $$

Lo cual se cumple si y solo si todos los valores son iguales a la media, por lo que probamos que la proposición es verdadera.

### Demostración por contradicción

Para probar una proposición $P$ por contradicción se debe asumir que $P$ es falsa y luego realizar una serie de deducciones lógicas que nos lleven a que una proposición verdadera conocida es falsa.

Ejemplo:

$$ \sqrt{2} \text{ es irracional}$$

Asumamos que $\sqrt{2}$ es racional, lo que implica que existen enteros $a$ y $b$ tales que:

$$ \sqrt{2} = \frac{a}{b} $$

Más aún, asumamos que $a$ y $b$ forman una fracción irreductible, por lo que $mcd(a,b) = 1$. 

Si elevamos al cuadrado ambas partes de la ecuación obtenemos que:

$$ 2 = \frac{a^{2}}{b^{2}} \rightarrow 2b^{2} = a^{2} $$

Eso quiere decir que $a^{2}$ es un múltiplo de $2$, y como $2$ es primo, deducimos que $a$ es múltiplo de $2$ también.

Ahora, como $a$ es múltiplo de $2$, entonces $a^{2}$ es múltiplo de $4$. Si $a^{2} = 4k$, reemplazando:

$$ 2b^{2} = 4k \rightarrow b^{2} = 2k $$

De manera análoga al caso de $a^{2}$ siendo múltiplo de $2$, deducimos que $b$ es múltiplo de $2$.

Concluimos que $a$ y $b$ son múltiplos de $2$ al mismo tiempo, cuando en un primer momento señalamos que no tenían factores en común: Llegamos a una contradicción, por lo que la proposición es verdadera.

## Algoritmos: Búsqueda Completa

La Búsqueda Completa es un paradigma de resolución de problemas en el cual se analiza todo el universo de posibilidades de candidatos de una instancia de problema.

A pesar de que el concepto en sí intente analizar todos los candidatos a solución, la meta es poder reducir razonablemente el conjunto a analizar para no usar tiempo de más. 

Como técnica, es muy útil para determinar patrones de comportamiento de algunos problemas a cuya solución no es fácil llegar, así como para poder realizar pruebas sobre instancias pequeñas de un problema en el que claramente una solución con este paradigma daría `TLE` y comparar las respuestas con una solución más eficiente.

Para recorrer algunas etapas del paradigma, usaremos como ejemplo básico el hallar los factores primos de un número $n$.

### Analizando todas las soluciones

Para poder determinar los factores primos de un número de manera natural, bastaría con verificar qué números entre $1$ y $n$ son divisores de $n$ y además son primos.

Esto nos daría una idea así:

```Python
for i in range(1, n+1):
    if n % i == 0:
        if i in primes:
            answer.add(i)
```

Este algoritmo nos daría una complejidad de $O(n + d\cdot p)$, donde $p$ es la complejidad de verificar si un número es primo o no y $d$ es la cantidad de divisores de $n$. El verificar si un número es primo o no de manera eficiente queda como ejercicio para el lector.

### Reduciendo el espacio de búsqueda

Para poder reducir el espacio de búsqueda deberemos realizar un análisis más profundo de las características del problema, por lo que debemos notar que:

$$ d | n \text{ y } d \leq \sqrt{n} \longrightarrow \frac{n}{d} \geq \sqrt{n} $$

Lo cual es fácil de probar basándonos en la desigualdad de la proposición inicial:

$$ d \leq \sqrt{n} $$

$$ 1 \leq \frac{\sqrt{n}}{d} $$

$$ \sqrt{n} \leq \sqrt{n}\cdot \frac{\sqrt{n}}{d} = \frac{n}{d} $$

Lo que nos quiere decir lo anterior es que si tenemos un divisor de $n$ al lado izquierdo de la raíz, entonces el divisor generado por $\frac{n}{d}$ está al lado derecho, siendo el caso $d = \sqrt{n}$ el único en el que ambos están en el mismo punto.

Lo anterior nos sirve como referencia para realizar la siguiente proposición:

$$ \exists\text{ } p | n \text{ con }p\text{ el mínimo primo posible y } p < n \longrightarrow p \leq \sqrt{n} $$

Lo cual no es muy difícil de probar:

$$ p \leq \frac{n}{p} \rightarrow p^{2} \leq n \rightarrow p \leq \sqrt{n} $$

La primera desigualdad se concluye debido a que $p$ siendo el menor primo implica que si dividimos $n$ entre $p$, entre sus factores quedarán $p$ (si es que su exponente no es 1) o sino algún otro primo mayor que $p$ (por condición de $p$).

Dada esta observación, notamos que si $n$ es un número compuesto, entonces su factor primo más pequeño será menor o igual a su raíz cuadrada. Otro dato más es que si usamos la contrapositiva respecto a la existencia de $p$, concluimos que si no encontramos algún $p$ en el rango $\left[1,\sqrt{n}\right]$ entonces no existe tal $p$, por lo que $n$ sería primo.

Finalmente, podemos definir $f(n)$ como el conjunto de factores primos de $n$, llegaremos a que:

$$ f(n) = \{p\} \cup f\left(\frac{n}{p^{a}}\right) $$

Donde $a$ es la máxima potencia tal que $p^{a} | n$. La igualdad de los conjuntos es evidente, pues $\frac{n}{p^{a}}$ tiene los factores de $n$ que no sean $p$.

Usando esta función $f$, podemos definir nuestro algoritmo optimizado:

```Python
for i in range(1, isqrt(n)+1):
    if n % i == 0:
        if i in primes:
            while n % i == 0:
                n = n // i
            answer.add(i)

if n != 1:
    answer.add(n)
```

La primera línea del algoritmo propuesto usa la cota optimizada que probamos en los pasos anteriores, el bucle interno de while usa la función $f$ como base para reducir el $n$ como corresponde y la última condición solamente verifica que el $n$ resultante sea agregado si es un número primo.

La complejidad obtenida ahora es $O(\sqrt{n}\cdot p)$, donde $p$ es la complejidad de verificar si un número es primo o no.

Podemos romper el límite en este caso y dar la siguiente observación:

$$ S(n)\text{ son los divisores de }n\text{ ordenados ascendentemente} \longrightarrow S(n)_{2}\text{ es el menor primo que divide a }n $$

Esta proposición se puede probar por contradicción, la prueba se deja como ejercicio para el lector (se dará en clase también). Finalmente notamos que deseamos buscar siempre el $S(n)_{2}$ en nuestro algoritmo.

Con la proposición anterior demostrada, llegamos al algoritmo final:

```Python
for i in range(2, isqrt(n)+1):
    if n % i == 0:
        while n % i == 0:
            n = n // i
        answer.add(i)
if n != 1:
    answer.add(n)
```

La modificación en la primera línea se da debido a que el $1$ siempre dividirá a $n$ pero no es primo; además, por la prueba anterior, no puede ser $S(n)_{2}$. Notemos que hemos quitado la línea de verificación de primo de $i$ también por la proposición hecha.

Al finalizar todo el análisis, obtuvimos un algoritmo con complejidad $O(\sqrt{n})$.
